# Appendix 10.3: Search & Retrieval

Did you know you can use Claude to **search through Wikipedia, the web, or your own documents for you**? Claude can find and retrieve information, at which point you can also use Claude to summarize and synthesize it, write novel content from what it found, and much more.

**The easiest path today: the native `web_search` server tool.** Add `{"type": "web_search_20250305", "name": "web_search"}` to your `tools` list (same `tools` parameter you used in Appendix 10.2) and Claude will search the web and cite its sources automatically — no scraping, no API keys, no manual result-parsing. This is now the default way to give Claude live, up-to-date knowledge, and it pairs naturally with the native Citations feature in Appendix 10.7 so every claim is traceable back to a source.

**For your own private documents** (support tickets, internal wikis, proprietary datasets), you still want **Retrieval-Augmented Generation (RAG)**: embed your documents in a vector database, retrieve the most relevant chunks for a given query, and pass those chunks to Claude as context (optionally as `document` content blocks so you get citations for free). See the [RAG cookbook examples](https://github.com/anthropics/anthropic-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb) to learn how to supplement Claude's knowledge with data retrieved from vector databases, Wikipedia, the internet, and more, including how to use [embeddings](https://docs.anthropic.com/claude/docs/embeddings) and vector database tools.

For long-document use cases, also keep in mind that context windows have grown enormously since this course was first written — for many RAG use cases today, "just put the whole document (or many documents) in context" is a viable and much simpler alternative to a full vector-retrieval pipeline. Combine that with Prompt Caching (Appendix 10.4) so you're not paying to re-process the same large context on every query.
